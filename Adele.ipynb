### Findings from above correlation table:

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop([' Sales Transaction','Number of Vehicles','Distance to Store (KM)','Family Size'],axis=1)    #our independent variables
y = df[' Sales Transaction']                #our dependent variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#allocate the method to an object:
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
accuracy = regressor.score(X_test, y_test)
print('Accuracy = '+ str(accuracy))
```

```python
print(regressor.intercept_)  
```

```python
print(regressor.coef_)
```
